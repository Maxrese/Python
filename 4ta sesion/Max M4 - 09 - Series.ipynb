# Ejercicio 5

Usar operadores relacionales para comparar los objetos Series dados.



````python
serie1 = pd.Series([1, 2, 3, 5, 7])
serie2 = pd.Series([13, 17, 11, 5, 3])
````

````python
menor_que = serie1 < serie2
menor_igual_que = serie1 <= serie2
mayor_que = serie1 > serie2
mayor_igual_que = serie1 >= serie2
igual_a = serie1 == serie2
distinto_de = serie1 != serie2
print("Menor que:", menor_que)
print("Menor o igual que:", menor_igual_que)
print("Mayor que:", mayor_que)
print("Mayor o igual que:", mayor_igual_que)
print("Igual a:", igual_a)
print("Distinto de:", distinto_de)
````

¿Qué ocurre? ¿Se comparan todos los elementos? ¿Se compara el objeto únicamente...?



```
Los operadores relacionales comparan los elementos de las series elemento a elemento y devuelven una nueva serie booleana con los resultados de las comparaciones.
```

PISTA: Podrías buscar en internet how to change index order series python



````python
datos = [1, 2, 3, 4, 5]

indices = ['A', 'B', 'C', 'D', 'E']

serie = pd.Series(data=datos, index=indices)

serie
````

# Ejercicio 21 (Opcional)

Operaciones aritméticas con métodos de la clase Series: `add`, `subtract`, `multiply`, y `divide`.



````python
serie1 = pd.Series(list(range(1, 11)))

serie1
````

````python
serie2 = pd.Series(list(range(10, 0, -1)))

serie2
````

````python
serie1 = pd.Series([1, 2, 3, 4, 5])
serie2 = pd.Series([6, 7, 8, 9, 10])
suma = serie1.add(serie2)
resta = serie1.subtract(serie2)
multiplicacion = serie1.multiply(serie2)
division = serie1.divide(serie2)
print("Suma:")
print(suma)
print("Resta:")
print(resta)
print("Multiplicación:")
print(multiplicacion)
print("División:")
print(division)
````
