## Ejercicio 1

Contar cuántos valores hay en una lista de enteros. Contar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
contador = 0
#Respuesta:
#Debe dar 5
for valor in nums:
    contador += 1
    if valor == 0:
        break

print(contador)
```

## Ejercicio 2

Sumar los valores que hay en una lista de enteros. Sumar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
suma = 0
#Respuesta:
#Debe dar 21

for valor in nums:
    if valor == 0:
        break
    suma += valor
print(suma)
```

## Ejercicio 3

Calcular el valor medio de los valores que hay en una lista de enteros. Calcular solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
suma = 0
contador = 0
#Respuesta:
#Debe dar 4.2
sublista =[]
for valor in nums:
    sublista.append(valor)
    if valor == 0:
        break

suma = sum(sublista)
longitud = len(sublista)
valor_medio = suma / longitud
print("La media da", valor_medio)
```

## Ejercicio 5 (Opcional)

Crear una lista a partir de una secuencia de valores de entrada terminada en *.
Una vez creada, modificarla elevando al cuadrado todos los pares y al cubo todos los impares, y mostrarla.

```python
lista = [3,56,6,7,"*",2,78]
#Respuesta:
#Debe dar [27, 3136, 36, 343]
nueva_lista = []

for elemento in lista:
    if isinstance(elemento, int):
        if elemento % 2 == 0:
            nueva_lista.append(elemento ** 2)
        else:
            nueva_lista.append(elemento ** 3)
    else:
        break

print(nueva_lista)
```
